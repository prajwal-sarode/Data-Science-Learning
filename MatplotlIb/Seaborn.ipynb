```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Pokemon.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
sns.lmplot(x='Attack',y='Defense', data=data)
plt.show()
# MORDERATE POSITIVE CORELATION
```

```python
sns.lmplot(x='Attack',y='Defense', data=data,fit_reg=False)
plt.show()
```

```python
sns.lmplot(x='Attack',y='Defense', data=data,fit_reg=False,hue='Legendary')
plt.show()
```

```python
data['Legendary'].value_counts()
```

```python
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])

x = np.arange(1,10)
a1.plot(x,np.exp(x),'b')

a1.set_title('range of number')

# set limit for x and y axis
plt.ylim(0,10000)
plt.xlim(0,10)

# set labels
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
data.head()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=data)
plt.show()
```

```python
data.describe().T
```

```python
data = data.drop(['#','Total','Legendary','Generation'],axis=1)
data.head()
```

```python
plt.figure(figsize=(15,15))
sns.boxplot(data=data)
plt.show()
```

```python
corr = data.corr() #calculating correlation
corr
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr) #creating HeatMap
plt.show()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(corr,annot=True)
plt.show()
```

 # Univariate Visualisation

## DistPlot

```python
sns.distplot(data['Defense'])
plt.show()
```

```python
sns.boxplot(data['Defense'])
plt.show()
```

```python
sns.distplot(data['Defense'],kde=False)
plt.show()
```

```python
sns.histplot(data['Defense'],kde=True)
plt.show()
```

```python
sns.kdeplot(data['Defense'])
plt.show()
```

# jointplot

```python
sns.jointplot(data['Defense'],data['Attack'])
plt.show()
```

```python
sns.jointplot(data['Defense'],data['Attack'],kind='kde')
plt.show()
```

```python
sns.jointplot(data['Defense'],data['Attack'], kind='reg')
plt.show()
#good when you want to explain the residuals
```

```python
sns.pairplot(data[['Defense','Attack','HP']],kind='scatter')
plt.show()
```
