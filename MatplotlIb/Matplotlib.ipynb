```python
x = np.array([2,4,6,8,10,12,14,16])
y = x/2

plt.figure(figsize=(10,5)) # size of the fig
plt.scatter(x,y,c='green')
plt.show()
```

# Label the axes

```python
plt.plot([1,2,3,4])
plt.ylabel('vertical')
plt.xlabel('horizontal')
plt.show()
```

# Create multiple plots with subplots

```python
classes= ['john','sanjay','ellen','sandra','nancy','shubham','philip']
score=[10,15,20,30,40,60,100]

plt.figure(figsize=(20,10))

plt.subplot(2,3,3)
plt.bar(classes, score)

plt.subplot(2,3,2)
plt.scatter(classes, score)

plt.subplot(2,3,1)
plt.plot(classes, score)

plt.subplot(2,3,4)
plt.bar(classes, score)

plt.suptitle('Categorical Plotting')
plt.show()
```
